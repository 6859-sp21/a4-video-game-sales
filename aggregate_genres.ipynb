```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("vgsales_clean_2.csv")
```

```python
data
```

```python
data = data.groupby(level = ["Genre","Year"]).agg({"NA_Sales": sum, "EU_Sales": sum, "JP_Sales": sum, "Other_Sales": sum, "Global_Sales": sum})
```

```python
data.columns
```

```python
genres = set()
for (genre, year), row in data.iterrows():
    genres.add(genre)
genres = list(genres)
```

```python
out_data = {}
for (genre, year), row in data.iterrows():
    if year in out_data:
        out_data[year][genre] = row["Global_Sales"]
    else:
        out_data[year] = {}
        out_data[year][genre] = row["Global_Sales"]
    
```

```python
X = []
for year in out_data:
    x = [year]
    d = out_data[year]
    for genre in genres:
        if genre in d:
            x.append(d[genre])
        else:
            x.append(0)
    X.append(x)

        
    
```

```python
out = pd.DataFrame(X, columns = ["year"] + list(genres))
```

```python
out.to_csv("aggregated_genre.csv")
```
